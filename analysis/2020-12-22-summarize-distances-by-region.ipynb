# Summarize distances by titers and mutations

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

%matplotlib inline
```

```python
df = pd.read_table("../auspice_tables/flu_seasonal_h3n2_ha_21y_global.tsv")
```

```python
na_df = pd.read_table("../auspice_tables/flu_seasonal_h3n2_ha_21y_north-america.tsv")
```

```python
fig, ax = plt.subplots(1, 1, figsize=(6, 6))
ax.plot(
    df["ep"],
    df["cTiterSub"],
    "o",
    alpha=0.5
)
ax.set_xlabel("ep")
ax.set_ylabel("cTiterSub")
```

```python
fig, ax = plt.subplots(1, 1, figsize=(6, 6))
ax.plot(
    df["ep"],
    df["cTiter"],
    "o",
    alpha=0.5
)
ax.set_xlabel("ep")
ax.set_ylabel("cTiter")
```

```python
fig, ax = plt.subplots(1, 1, figsize=(6, 6))
ax.plot(
    df["cTiter"],
    df["cTiterSub"],
    "o",
    alpha=0.5
)
ax.set_xlabel("cTiter")
ax.set_ylabel("cTiterSub")
```

```python
fig, ax = plt.subplots(1, 1, figsize=(6, 6))
ax.plot(
    df["cTiter_seasonal"],
    df["cTiterSub_seasonal"],
    "o",
    alpha=0.5
)
ax.set_xlabel("cTiter_seasonal")
ax.set_ylabel("cTiterSub_seasonal")
```

```python
fig, ax = plt.subplots(1, 1, figsize=(6, 6))
ax.plot(
    df["cTiter_seasonal_vaccine"],
    df["cTiterSub_seasonal_vaccine"],
    "o",
    alpha=0.5
)
ax.set_xlabel("cTiter_seasonal_vaccine")
ax.set_ylabel("cTiterSub_seasonal_vaccine")
```

```python
fig, ax = plt.subplots(1, 1, figsize=(6, 6))
ax.plot(
    df["cTiter_seasonal"],
    df["cTiter_seasonal_vaccine"],
    "o",
    alpha=0.5
)
ax.set_xlabel("cTiter_seasonal")
ax.set_ylabel("cTiter_seasonal_vaccine")
```

```python
fig, ax = plt.subplots(1, 1, figsize=(6, 6))
ax.plot(
    df["cTiterSub_seasonal"],
    df["cTiterSub_seasonal_vaccine"],
    "o",
    alpha=0.5
)
ax.set_xlabel("cTiterSub_seasonal")
ax.set_ylabel("cTiterSub_seasonal_vaccine")
```

```python
fig, axes = plt.subplots(2, 1, figsize=(6, 6), sharex=True, sharey=True)
bins = np.linspace(0, 10)
axes[0].hist(
    df["ep"],
    bins=bins
)
axes[0].set_title("Global")

axes[1].hist(
    na_df["ep"],
    bins=bins
)
axes[1].set_title("North America")
axes[1].set_xlabel("ep")
```
